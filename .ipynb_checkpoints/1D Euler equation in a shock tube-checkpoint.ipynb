# Solving Euler's equation for compressible flow in a shock tube

Euler's equation describes the behaviour of compressible, inviscid flow. In one dimension, the conservative form of Euler's equation is written for a perfect (ideal) gas as

$$ \dfrac{\partial \mathbf{U}}{\partial t} + \dfrac{\partial \mathbf{F}}{\partial x} = 0 $$

where $\mathbf{U}$ and $\mathbf{F}$ are

$$ \mathbf{U} = \left[
\begin{array}{c}
\rho \\
\rho u \\
\rho E
\end{array}\right] \quad 
\mathbf{F} = \left[
\begin{array}{c}
\rho u \\
\rho u^2 + p \\
\rho u (E + p)
\end{array}\right]. $$

Here, $E = \frac{p}{\gamma - 1} + \frac{1}{2}\rho u^2$ is the specific internal energy and $\gamma$ is the specific heat ratio. For this exercise, $\gamma = 1.4$ is used. $\mathbf{U}$ represents the vector of conserved quantities (mass, momentum, energy), while $\mathbf{F}$ represents the flux quantities.

Euler's equation can be solved via a number of numerical schemes. For simplicity, we will consider only one here: the MacCormack method.

## MacCormack method

The MacCormack scheme is a two-step scheme defined by

$$ \text{Step 1: } \mathbf{U}_i^{*} = \mathbf{U}_i^n - \dfrac{\Delta t}{\Delta x} ( \mathbf{F}_{i+1}^n - \mathbf{F}_{i}^n ) $$
$$ \text{Step 2: } \mathbf{U}_i^{n+1} = \dfrac{1}{2}\left[ \mathbf{U}_i^n + \mathbf{U}_i^{*} \right] - \dfrac{\Delta t}{2\Delta x} ( \mathbf{F}_{i}^* - \mathbf{F}_{i-1}^*) $$

Note, that in the corrector step, the fluxes are evaluated using variables from the corrector step, i.e. $\mathbf{F}(\mathbf{U}^*)$.

The predictor step employs a first-order forwards difference, while the corrector step employs a backwards differences. Therefore, either step will be unstable to one characteristic speed, but the overall scheme is stable and second-order accurate for the linear case. For the case of non-linear stability, the MacCormack method is known to suffer from a non-linear instability due to the lack of dissipation. 

## Problem description

We will solve the classic shock tube problem for the Euler equation. This problem is plotted below:



A shock tube consists of a cylinder with a membrane separating two gas volumes. By adjusting the conditions of the gases on either side of the membrane and then rupturing the membrane, a variety of gas dynamic conditions can be tested. This approach is used to generate ultra-high Mach numbers in experimental tests of spacecraft re-entry, for example. A real shock tube can be seen in this video:

## Numerical solution - Python code

```python
import numpy as np
import matplotlib.pyplot as plt

# Sub-routine that extracts primitive variables from the solution vector

def primitive(U):
    rho = U[:,0]
    u = U[:,1]/rho
    E = U[:,2]/rho
    p = (gamma-1)*(U[:,2]-0.5*rho*u**2)
    return rho,u,E,p

# Sub-routine that generates a flux vector from primitive variables

def flux(rho,u,p,E):
    FF = np.zeros((np.size(rho),3))
    FF[:,0] = rho*u
    FF[:,1] = rho*u**2 + p
    FF[:,2] = rho*u*E + p*u
    return FF

# Define spatial coordinates
xL = -1
xR = 1

# Define number of mesh points
n = 81

# Define spatial step size
dx = (xR-xL)/n

# Define initial conditions on the L (left) and R (right) side of the membrane
rhoL = 1.0
pL = 1.0
uL = 0
rhoR = 0.125
pR = 0.1
uR = 0

# Gas specific heat ratio
gamma = 1.4

# Define how long the solution will be computed for (in seconds)
tmax = 0.25 

# Initialize the solution and flux vectors
U = np.zeros((n,3))
FF = np.zeros((n,3))

U[0:int(n/2),0] = rhoL
U[int(n/2):,0] = rhoR
U[0:int(n/2),1] = rhoL*uL
U[int(n/2):,1] = rhoR*uR
U[0:int(n/2),2] = pL/(gamma-1) + 0.5*rhoL*uL**2
U[int(n/2):,2] = pR/(gamma-1) + 0.5*rhoR*uR**2

FF[0:int(n/2),0] = rhoL*uL
FF[int(n/2):,0] = rhoR*uR
FF[0:int(n/2),1] = rhoL*uL**2 + pL
FF[int(n/2):,1] = rhoR*uR**2 + pR
FF[0:int(n/2),2] = (pL/(gamma-1) + 0.5*rhoL*uL**2 + pL)*uL 
FF[int(n/2):,2] = (pR/(gamma-1) + 0.5*rhoR*uR**2 + pR)*uR

# Get primitive variable vectors from the solution vector.
[rho, u, E, p] = primitive(U)

# Define solution information
eps = 0.01  # - dissipation constant - used for smoothing the solution
t = 0       # - initializet time to zero
l = 0       # - this is a counter for counting the number of time steps
CFL = 1     # - CFL number - it is defined for specifying the time step size

# Solve in time
while (t < tmax):
    
    # Evaluate the time step size based on the user-specified CFL condition:
    # CFL = (|umax| + a) dx / dt for all cells
    
    amax = np.max((gamma*p/rho)**0.5)   # - max speed of sound in all cells
    umax = np.max(np.abs(u))            # - max absolute velocity in all cells
    dt = CFL*dx/(umax+amax)             # - define time step size 
    
    # Print an output message every ten time steps
    if (l%10 == 0):
        print('Step',l,' t=',t,'umax=', umax,'amax=',amax,'dt =', dt)
    
    
    # Evaluate predictor step
    Ubar = U.copy()
    Ubar[1:-1,:] = U[1:-1,:] - dt/dx*(FF[2:,:] - FF[1:-1,:])
    
    # Add smoothing to remove numerical dispersion error
    Ubar[2:-2,:] = Ubar[2:-2,:] - eps*(Ubar[4:,:] - 4*Ubar[3:-1,:] + 6*Ubar[2:-2,:] - 4*Ubar[1:-3,:] + Ubar[:-4,:])
    
    # Convert solution vector to primitive variables
    [rho, u, E, p] = primitive(Ubar)
    # Reconstruct flux vector FF using the updated primitive variables
    FFbar = flux(rho,u,p,E)
    
    # Evaluate corrector step
    U[1:-1,:] = 0.5*(U[1:-1,:] + Ubar[1:-1,:]) - 0.5*dt/dx*(FFbar[1:-1,:] - FFbar[:-2,:])
    
    # Add smoothing
    U[2:-2,:] = U[2:-2,:] - eps*(U[4:,:] - 4*U[3:-1,:] + 6*U[2:-2,:] - 4*U[1:-3,:] + U[:-4,:])
    
    # Update flux vector
    [rho, u, E, p] = primitive(U)
    FF = flux(rho, u, p, E)
    
    # Update time
    t += dt
    l += 1

print('Solution obtained in',l-1,'time steps.')
```

## Analytical solution - Python code

```python
def F(p,pL,pR,gamma,rhoL,rhoR,aL,aR):
    
    AL = 2/((gamma+1)*rhoL)
    BL = (gamma-1)/(gamma+1)*(pL)
    AR = 2/((gamma+1)*rhoR)
    BR = (gamma-1)/(gamma+1)*(pR)
    
    if (p > pL):  # shock
        FL = (p-pL) * (AL/(p+BL))**0.5
    else:  # rarefaction
        FL = 2*aL/(gamma-1)*((p/pL)**((gamma-1)/(2*gamma)) - 1)
    
    if (p > pR):  # shock
        FR = (p-pR) * (AR/(p+BR))**0.5
    else:  # rarefaction
        FR = 2*aR/(gamma-1)*((p/pR)**((gamma-1)/(2*gamma)) - 1)
    
    return FR,FL

def FPrime(p,pL,pR,gamma,rhoL,rhoR,aL,aR):
    
    AL = 2/((gamma+1)*rhoL)
    BL = (gamma-1)/(gamma+1)*(pL)
    AR = 2/((gamma+1)*rhoR)
    BR = (gamma-1)/(gamma+1)*(pR)
    
    if (p > pL):
        FPL = (AL/(BL+p))**0.5 * (1 - (p-pL)/(2*(BL+p)))
    else:
        FPL = 1/(rhoL*aL)*((p/pL)**(-1*(gamma+1)/(2*gamma)))
        
    if (p > pR):
        FPR = (AR/(BR+p))**0.5 * (1 - (p-pR)/(2*(BR+p)))
    else:
        FPR = 1/(rhoR*aR)*((p/pR)**(-1*(gamma+1)/(2*gamma)))

    return FPL, FPR    

def G(p,pL,pR,gamma,rhoL,rhoR):
    
    AL = 2/((gamma+1)*rhoL)
    BL = (gamma-1)/(gamma+1)*(pL)
    AR = 2/((gamma+1)*rhoR)
    BR = (gamma-1)/(gamma+1)*(pR)
    GL = (AL/(p+BL))**0.5
    GR = (AR/(p+BR))**0.5
    return GL, GR

def guess(pL,pR,uL,uR,rhoL,rhoR,aL,aR,gamma):
    
    # Below are four possible initial guesses for the pressure in the star region

    # # Case 1 - Mean variables
    # pMN = 0.5*(pL+pR)

    # Case 2 - Primitive variables
    pPV = 0.5*(pL+pR) - 1/8*(uR-uL)*(rhoL+rhoR)*(aL+aR)

    # Case 3 - Two shock solution
    GV = G(pPV,pL,pR,gamma,rhoL,rhoR)
    pTS = (GV[0]*pL + GV[1]*pR - (uR-uL))/(GV[0]+GV[1])

    # Case 4 - Two rarefaction wave solution
    pTR = ((aL + aR - 0.5*(gamma-1)*(uR-uL))/(aL/(pL**((gamma-1)/(2*gamma))) + aR/(pR**((gamma-1)/(2*gamma)))))**((2*gamma)/(gamma-1))

    pMAX = max([0,pPV])
    pMIN = min([pL,pR])
    QMAX = pMAX/pMIN
    
    if ( (QMAX <= 2.0) and (pMIN <= pPV) and (pPV <= pMAX) ):
        print('Using primitive variable initial guess for pressure')
        return pPV
    elif (pPV < pMIN):
        print('Using two-rarefaction wave initial guess for pressure')
        return pTR
    else:
        print('Using two-shock wave initial guess for pressure')
        return pTS
    
def solve(p,pL,pR,rhoL,rhoR,uL,uR,aL,aR,gamma):
    TOL = 1E-6    
    CHA = 1
    i = 0
    while (CHA > TOL):
        [FL, FR] = F(p,pL,pR,gamma,rhoL,rhoR,aL,aR)
        [FPL,FPR] = FPrime(p,pL,pR,gamma,rhoL,rhoR,aL,aR)
        pn = p - (FL + FR + (uR-uL))/(FPL+FPR)
        CHA = abs(pn-p)/(0.5*(pn+p))
        p = pn
        i+=1
    
    print('Star region pressure is',p,'. Reached in ',i,'iterations.')
    ustar = 0.5*(uL+uR) - 0.5*(FR-FL)
    print('Star region velocity is',ustar)
    return p, ustar

def sample(pstar,ustar,xL,xR,pL,pR,rhoL,rhoR,uL,uR,aL,aR,gamma,n,tmax):
    x = np.linspace(xL,xR,n)
    rho = np.zeros(n)
    u = np.zeros(n)
    p = np.zeros(n)
    e = np.zeros(n)
    
    for i in range(0,n):
        S = x[i]/tmax
        if (S < ustar): # left of the contact discontinuity
            if (pstar > pL): # left wave is a shock
                SL = uL - aL*((gamma+1)/(2*gamma)*(pstar/pL) + (gamma-1)/(2*gamma))**0.5
                if (S < SL):
                    p[i] = pL
                    u[i] = uL
                    rho[i] = rhoL
                else:
                    rho[i] = rhoL*(pstar/pL + (gamma-1)/(gamma+1))/(gamma-1)/((gamma-1)/(gamma+1)*pstar/pL+1)
                    p[i] = pstar
                    u[i] = ustar
            else:  # left wave is a fan
                SHL = uL-aL
                STL = ustar - aL*(pstar/pL)**((gamma-1)/(2*gamma))
                if (S < SHL):
                    p[i] = pL
                    u[i] = uL
                    rho[i] = rhoL
                elif (S > STL):
                    rho[i] = rhoL*(pstar/pL)**(1/gamma)
                    u[i] = ustar
                    p[i] = pstar
                else:
                    rho[i] = rhoL*( 2/(gamma+1) + (gamma-1)/((gamma+1)*aL)*(uL-S))**(2/(gamma-1))
                    u[i] = 2/(gamma+1)*(aL + (gamma-1)/2*uL + S)
                    p[i] = pL*( 2/(gamma+1) + (gamma-1)/((gamma+1)*aL)*(uL-S))**(2*gamma/(gamma-1))
        else: # right of the contact discontinuity
            if (pstar > pR): # right wave is a shock
                SR = uR + aR*((gamma+1)/(2*gamma)*(pstar/pR) + (gamma-1)/(2*gamma))**0.5
                if (S > SR):
                    p[i] = pR
                    rho[i] = rhoR
                    u[i] = uR
                else:
                    p[i] = pstar
                    u[i] = ustar
                    rho[i] = rhoR*( pstar/pR + (gamma-1)/(gamma+1))/( (gamma-1)/(gamma+1)*pstar/pR + 1)
            else: # right wave is a fan
                SHR = uR+aR
                STR = ustar + aR*(pstar/pR)**((gamma-1)/(2*gamma))
                if (S > SHR):
                    p[i] = pR
                    u[i] = uR
                    rho[i] = rhoR
                elif (S < STR):
                    p[i] = pstar
                    u[i] = ustar
                    rho[i] = rhoR*(pstar/rhoR)**(1/gamma)
                else:
                    rho[i] = rhoR*( 2/(gamma+1) - (gamma-1)/((gamma+1)*aR)*(uR-S))**(2/(gamma-1))
                    u[i] = 2/(gamma+1)*(-1.0*aR + (gamma-1)/2*uR + S)
                    p[i] = pR*( 2/(gamma+1) - (gamma-1)/((gamma+1)*aR)*(uR-S))**(2*gamma/(gamma-1))
        e[i] = p[i]/((gamma-1)*rho[i])
    return x,rho,p,u,e
       
# Speed of sound
aL = (gamma*pL/rhoL)**0.5
aR = (gamma*pR/rhoR)**0.5

# Select initial pressure guess
p0 = guess(pL,pR,uL,uR,rhoL,rhoR,aL,aR,gamma)

# Iteratively solve for preessure
[pstar, ustar] = solve(p0,pL,pR,rhoL,rhoR,uL,uR,aL,aR,gamma)

# Sample solution space
[x,rhoa,pa,ua,ea] = sample(pstar,ustar,xL,xR,pL,pR,rhoL,rhoR,uL,uR,aL,aR,gamma,n,tmax)   
    
```

## Plot solution

```python
fig, axs = plt.subplots(2, 2,figsize=(8,6),dpi=100)
axs[0, 0].plot(x, p, 'o', label='MacCormack')
axs[0, 0].plot(x, pa, '-', label='Analytical')
axs[0, 0].legend()
axs[0, 0].set_ylabel('Pressure')
axs[0, 0].grid()
axs[0, 0].set_xlim((-1,1))
axs[0, 1].plot(x, u, 'o', x, ua, '-')
axs[0, 1].set_ylabel('Velocity')
axs[0, 1].grid()
axs[0, 1].set_xlim((-1,1))
axs[1, 0].plot(x, rho, 'o', x, rhoa, '-')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('Density')
axs[1, 0].grid()
axs[1, 0].set_xlim((-1,1))
axs[1, 1].plot(x, E, 'o', x, ea+0.5*ua**2, '-')
axs[1, 1].set_ylabel('Specific internal energy')
axs[1, 1].set_xlabel('x')
axs[1, 1].grid()
axs[1, 1].set_xlim((-1,1))
fig.tight_layout(pad=1.5)    
```
