## Introduction

The 1D linear convection equation has a form

$$ \dfrac{\partial u}{\partial t} + c \dfrac{\partial u}{\partial x} = \epsilon \dfrac{\partial^2 u}{\partial x^2} $$

where $u = f(x,t)$ is the amplitude of the wave, $c$ is the convection speed of the wave in m/s, and $\epsilon$ is a damping factor due to dissipation of the wave as it travels. When $\epsilon = 0$, the wave is undamped and it will convect uniformly without damping. When $\epsilon > 0$, the wave amplitude will be damped as it convects. Naturally, the problem is driven by initial conditions, $u_0(x,0) = f(x)$.

We notice that the 1D linear convection equation is linear because the coefficient $c$ is a constant. If $c = f(x)$, the problem would become quasi-linear, and if $c = f(u)$, the problem would be non-linear. Non-linear convection is treated in a later exercise; we'll see that it involves some challenging and interesting features.

One key development in this problem is the addition of _unsteady_ effects, clearly evident in our governing by the fact there is a time derivative present. This gives us the opportunity to talk about solution of unsteady problems. Just as the spatial coordinates, time now must be discretized. There is a major difference between differentiating with respect to space and differentiating with respect to time, though: the directionality of influence. A force at any spatial location may influence the flow in any direction, but a force at any instant can only affect the flow in the future--there is no backwards influence. Unsteady flows are therefore parabolic-like in time, and thus no conditions can be placed on the solution (except at the boundaries) at any time after the calculation begins.

Let's consider the discretization of our 1D linear convection equation. We'll begin by discretizing the spatial derivatives. Let's use a 2nd-order centered difference for the spatial derivatives:

$$ \dfrac{\partial u}{\partial x} \approx \dfrac{u^{n}_{i+1} - u^{n}_{i-1}}{2\Delta x}, \quad \dfrac{\partial^2 u}{\partial x^2} \approx \dfrac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{\Delta x^2} $$

Now that the spatial derivatives are approximated, the governing equation becomes an ordinary differential equation in time:

$$ \dfrac{du}{dt} = \epsilon \dfrac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{\Delta x^2} - c \dfrac{u^{n}_{i+1} - u^{n}_{i-1}}{2\Delta x} $$
which we can simplify to 
$$ \dfrac{du}{dt} = f(t_n,u^n), $$

where 
$$ f(t_n,u^n) = \epsilon \dfrac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{\Delta x^2} - c \dfrac{u^{n}_{i+1} - u^{n}_{i-1}}{2\Delta x} $$

Discretization of this ODE can involve a number of different techniques, each with different properties. We will discuss them now.

### Two-level methods

Some simple procedures for discretizing the ODE are shown below:

**Explicit Euler:**
$$ u^{n+1} = u^n + f(t_n, u^n) \Delta t$$
It estimates the temporal derivative using a forwards difference, so it is also called the _forwards_ Euler method.

**Implicit Euler:**
$$ u^{n+1} = u^n + f(t_{n+1}, u^{n+1}) \Delta t$$
Because this method uses the new time to evaluate the RHS function, it is an implicit method, analogous to the line Gauss-Seidel method. It is sometimes referred to as the _backwards_ Euler method.

**Trapezoid rule:**
$$ u^{n+1} = u^n + \dfrac{1}{2} \left[ f(t_n,u^n) + f(t_{n+1},u^{n+1})\right] \Delta t$$
This method is an average of the explicit and implicit Euler method, and is the basis for the Crank-Nicholson method.

The above methods are all called two-level methods because they involve time at two distinct time levels $t_n$ and $t_{n+1} = t_n + \Delta t$. We can show that the explicit and implicit Euler methods are first-order accurate in time, that is the discretization error scales as $\mathcal{O}(\Delta t)$. The trapezoid rule is second-order accurate, with discretization error scaling as $\mathcal{O}[(\Delta t)^2]$. The explicit method is easy to program, requires less computer memory, and incurs few calculations per time step, but may be unstable if $\Delta t$ is too large. On the other hand, the implicit and trapezoid methods are harder to program, consume more memory, and require more calculations per time step but are far more stable.

**Predictor-corrector methods.** One may ask: is it possible to combine the benefits of both explicit methods (speed, simplicity) and implicit methods (stability)? The answer is yes; such schemes are called _predictor-corrector methods_. The most common uses explict Euler method to predict the new time level,

$$\text{Predictor:} \quad u^{*\;n+1} = u^n + f(t_n, u^n)\Delta t,$$

where the $*$ indicates this is the predicted solution at $t_{n+1}$. Next, the solution is corrected by applying trapezoid rule using $u^{*\; n+1}$ to compute the derivative:

$$\text{Corrector:} \quad u^{n+1} = u^n + \dfrac{1}{2}\left[ f(t_n, u^n) + f(t_{n+1},u^{*\;n+1})\right]\Delta t.$$

This method is second-order accurate but has roughly the same stability bounds as the explicit Euler.

### Higher-order methods

The two-point methods are limited to second-order accuracy in time. To increase the accuracy further, we need additional time levels (such as $t_{n-1}$, $t_{n-2}$, etc.) or intermediate time levels between $t_n$ and $t_{n+1}$. The former approach lead to _multipoint methods_ , while the latter approach is the _Runge-Kutta_ method.

**Adams-Bashforth method:** The best known multipoint method are the Adams family of methods. If a Lagrange polynomial is fit to $f(t_{n-m},u^{n-m})$, $f(t_{n-m+1},u^{n-m+1})$, $...$, $f(t_n,u^n)$, we obtain an explicit method of order $m+1$. A third-order Adams-Bashforth method is:

$$u^{n+1} = u^n + \dfrac{\Delta t}{12} \left[ 23 f(t_n,u^n) - 16 f(t_{n-1},u^{n-1}) + 5 f(t_{n-2},u^{n-2}) \right]. $$

**Adams-Moulton method:** If data at $t_{n+1}$ is included in the interpolation polynomial, the implicit Adams-Moulton method is obtained. A third-order method is:

$$u^{n+1} = u^n + \dfrac{\Delta t}{12} \left[ 5 f(t_{n+1},u^{n+1}) + 8 f(t_{n},u^{n}) - f(t_{n-1},u^{n-1}) \right]. $$

**Runge-Kutta methods:** The obvious disadvantage of multipoint methods is the additional memory required to store the old time steps. Runge-Kutta (RK) methods can achieve higher order without this overhead. There are numerous types of RK methods; the classical fourth-order is the RK4:

$$u^{*\; n+\frac{1}{2}} = u^n + \dfrac{\Delta t}{2} f(t_n,u^n)$$
$$u^{**\; n+\frac{1}{2}} = u^n + \dfrac{\Delta t}{2} f(t_{n+\frac{1}{2}},u^{*\;n+\frac{1}{2}})$$
$$u^{*\;n+1} = u^n + \Delta t f(t_{n+\frac{1}{2}}, u^{**\; n+\frac{1}{2}}) $$

$$u^{n+1} = u^n + \dfrac{\Delta t}{6} \left [ f(t_n,u^n) + 2 f(t_{n+\frac{1}{2}},u^{*\;n+\frac{1}{2}}) + 2 f(t_{n+\frac{1}{2}},u^{**\;n+\frac{1}{2}}) + f(t_{n+1},u^{*\;n+1})\right] $$
RK methods of a given order are more expensive than multipoint methods of comparable order. Howevever, RK methods are typically more accurate and stable.

## Problem statement

Solve the 1D linear convection equation for a periodic domain $x \in (-5,5)$ with an initial condition

$$u_0 = e^{-x^2}$$

The wave propagates at a convection speed of $c = 1$. Let's consider two cases: undamped case with $\epsilon = 0$ and damped case with $\epsilon = 5\times 10^{-4}$.

```python
import numpy as np
import matplotlib.pyplot as plt

nx = 101
xmin = -5
xmax = 5
x = np.linspace(xmin,xmax,nx)
dx = (xmax-xmin)/nx
c = 1
```

## Sub-routines

```python
def func(u,epsilon):
    # Sub-routine to evaluate the RHS function. Uses 2nd-order centered differences to discretize the spatial derivatives
    f = 0*u.copy()
    f[0] = epsilon*(u[1]-2*u[0]+u[-2])/(dt**2) - (u[1]-u[-2])/(2*dx)
    f[1:-1] = epsilon*(u[2:]-2*u[1:-1]+u[:-2])/(dt**2) - (u[2:]-u[:-2])/(2*dx)
    f[-1] = epsilon*(u[1]-2*u[-1]+u[-2])/(dt**2) - (u[1]-u[-2])/(2*dx)
    return f

def L2norm(u):
    # Sub-routine to evaluate the L2 norm of the error
    return (np.sum((u-np.exp(-x**2))**2))**0.5
```

## Explicit Euler method

```python
# Initial condition
u = np.exp(-x**2)

# Define time step
CFL = 0.2
dt = CFL*dx/c

## Undamped case
epsilon = 0

t = 0
tmax = 10 # This is the time required for the wave to convect back to its original position
plt.plot(x,u,label='$t=0$')
while (t < tmax):
    un = u.copy()
    u = un + dt*func(un,epsilon)
    t+=dt
plt.plot(x,u,label='$\epsilon = 0$')
plt.title('Explicit Euler')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')

print('Undamped L2norm(error) =',L2norm(u))

## Damped case
epsilon = 5e-4

# Initial condition
u = np.exp(-x**2)

t = 0
while (t < tmax):
    un = u.copy()
    u = un + dt*func(un,epsilon)
    t+=dt
plt.plot(x,u,label='$\epsilon = 5 \\times 10^{-4}$')
plt.legend()

print('Damped L2norm(error) =',L2norm(u))
```

Notice how for the undamped case, there is overshoot of the peak and a lagging phase error (the wiggle on the left side of the convected wave). We'll discuss the source of this error below. For the damped case, the overshoot and phase error have been removed by the viscous damping. This is reflected in the reduction in the $L^2$ norm of the error.

## Predictor-corrector method

```python
# Undamped case
u = np.exp(-x**2)
t = 0
epsilon = 0
plt.plot(x,u,label='$t=0$')
while (t < tmax):
    un = u.copy()
    ustar = un + dt*func(un,epsilon)                         # predictor
    u = un + dt*0.5*(func(un,epsilon) + func(ustar,epsilon)) # corrector
    t+=dt
plt.plot(x,u,label='$\epsilon = 0$')
print('Undamped L2norm(error) =',L2norm(u))

## Damped case
u = np.exp(-x**2)
t = 0
epsilon = 5e-4
while (t < tmax):
    un = u.copy()
    ustar = un + dt*func(un,epsilon)                         # predictor
    u = un + dt*0.5*(func(un,epsilon) + func(ustar,epsilon)) # corrector
    t+=dt
plt.plot(x,u,label='$\epsilon = 5\\times 10^{-4}$')
plt.title('Predictor-corrector')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Damped L2norm(error) =',L2norm(u))    
```

The predictor-corrector solution is much better than the explicit Euler method. For the undamped case, there is not overshoot of the wave peak and there are no high-frequency wiggles on the lagging side of the wave. We do see some error associated with the wave at locations of small amplitude (i.e. near $x$ = -2 and +2), where the numerical solution lags behind the exact wave. When damping is turned on, the wave begins to dissipate, resulting in a broading of the wave and a reduction in its amplitude.

## Adams-Bashforth method

```python
u = np.exp(-x**2)
plt.plot(x,u,label='$t=0$')

#Undamped case
epsilon = 0
t = 0
n = 0
while (t < tmax):
    # For the first time step, assign u_{n-1} = u_{n-2} = u0. 
    # For subsequent time steps, assign u_{n-2} = u_{n-1} and u_{n-1} = u_n from prior time step
    # assign u_{n-2} to u_{n-1} from the prior time step
    if (n == 0):
        un2 = u.copy() #u_{n-2}
        un1 = u.copy() #u_{n-1}
    else:
        un2 = un1.copy() #u_{n-2}
        un1 = un.copy() #u_{n-1}
    un = u.copy()   #u_n
    u = un + dt/12.0*(23.0*func(un,epsilon) - 16.0*func(un1,epsilon) + 5.0*func(un2,epsilon) )
    t+=dt
    n+=1

plt.plot(x,u,label='$\epsilon = 0$')
print('Undamped L2norm(error) =',L2norm(u))

## Damped case
u = np.exp(-x**2)
epsilon = 5e-4
t = 0
n = 0
while (t < tmax):
    if (n == 0):
        un2 = u.copy() #u_{n-2}
        un1 = u.copy() #u_{n-1}
    else:
        un2 = un1.copy() #u_{n-2}
        un1 = un.copy() #u_{n-1}
    un = u.copy()   #u_n
    u = un + dt/12.0*(23.0*func(un,epsilon) - 16.0*func(un1,epsilon) + 5.0*func(un2,epsilon) )
    t+=dt
    n+=1
plt.plot(x,u,label='$\epsilon = 5\\times 10^{-4}$')
plt.title('Adams-Bashforth')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Damped L2norm(error) =',L2norm(u))    
```

The solution accuracy is very similar to the predictor-corrector method, with modest phase error but little dissipation error for the undamped case, and dissipation with the damped case.

## Runge-Kutta method

```python
u = np.exp(-x**2)
plt.plot(x,u,label='$t=0$')

#Undamped case
epsilon = 0
t = 0
n = 0
while (t < tmax):
    un = u.copy()
    u1 = un + 0.5*dt*func(un,epsilon) # Step 1
    u2 = un + 0.5*dt*func(u1,epsilon) # Step 2
    u3 = un + dt*func(u2,epsilon)     # Step 3
    u = un + (1.0/6.0)*dt*(func(un,epsilon) + 2.0*func(u1,epsilon) + 2.0*func(u2,epsilon) + func(u3,epsilon))  # Step 4
    t+=dt

plt.plot(x,u,label='$\epsilon = 0$')
print('Undamped L2norm(error) =',L2norm(u))

## Damped case
u = np.exp(-x**2)
epsilon = 5e-4
t = 0
n = 0
while (t < tmax):
    un = u.copy()
    u1 = un + 0.5*dt*func(un,epsilon) # Step 1
    u2 = un + 0.5*dt*func(u1,epsilon) # Step 2
    u3 = un + dt*func(u2,epsilon)     # Step 3
    u = un + (1.0/6.0)*dt*(func(un,epsilon) + 2.0*func(u1,epsilon) + 2.0*func(u2,epsilon) + func(u3,epsilon))  # Step 4
    t+=dt
plt.plot(x,u,label='$\epsilon = 5\\times 10^{-4}$')
plt.title('Runge-Kutta')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Damped L2norm(error) =',L2norm(u))   
```

Again, the error for the Runge-Kutta is very similar as for the Adams-Bashforth and predictor-corrector method.

### A more in-depth look at the error

We see from the above results that the Runge-Kutta, Adams-Bashforth, and predictor-corrector methods perform similarly well. Let's investigate the numerical error a bit more in-depth. For this exercise, we'll focus on the undamped RK case. Let's allow the wave to convect for longer, say 5 cycles through our periodic domain. This corresponds to a max solution time of 50 s.

```python
u = np.exp(-x**2)
plt.plot(x,u,label='$t=0$')

tmax = 50
epsilon = 0
t = 0
n = 0
while (t < tmax):
    un = u.copy()
    u1 = un + 0.5*dt*func(un,epsilon) # Step 1
    u2 = un + 0.5*dt*func(u1,epsilon) # Step 2
    u3 = un + dt*func(u2,epsilon)     # Step 3
    u = un + (1.0/6.0)*dt*(func(un,epsilon) + 2.0*func(u1,epsilon) + 2.0*func(u2,epsilon) + func(u3,epsilon))  # Step 4
    t+=dt

plt.plot(x,u,label='$t = 50$')
plt.title('Runge-Kutta, $\epsilon = 0$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Undamped L2norm(error) =',L2norm(u))
```

We see that the phase error accummulates with the longer simulation time. By 50 s, the $L^2$ error norm has increased by about a factor of 5 compared to the error at 10 s. If we compute for even longer (say, 20 cycles), the phase error grows even larger and begins to completely distort the solution:

```python
u = np.exp(-x**2)
plt.plot(x,u,label='$t=0$')

tmax = 200
epsilon = 0
t = 0
n = 0
while (t < tmax):
    un = u.copy()
    u1 = un + 0.5*dt*func(un,epsilon) # Step 1
    u2 = un + 0.5*dt*func(u1,epsilon) # Step 2
    u3 = un + dt*func(u2,epsilon)     # Step 3
    u = un + (1.0/6.0)*dt*(func(un,epsilon) + 2.0*func(u1,epsilon) + 2.0*func(u2,epsilon) + func(u3,epsilon))  # Step 4
    t+=dt

plt.plot(x,u,label='$t = 200$')
plt.title('Runge-Kutta, $\epsilon = 0$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Undamped L2norm(error) =',L2norm(u))
```

Where does this phase error come from, and how can we manage it? Let's consider the RHS function $f(t_n, u^n)$:

$$ f(t_n,u^n) = \epsilon \dfrac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{\Delta x^2} - c \dfrac{u^{n}_{i+1} - u^{n}_{i-1}}{2\Delta x}. $$

The spatial derivatives in this function are approximated using 2nd-order centered finite differences. Let's try switching to a first-order backwards difference for the convection term (the first derivative),

$$ f(t_n,u^n) = \epsilon \dfrac{u_{i+1}^n - 2u_i^n + u_{i-1}^n}{\Delta x^2} - c \dfrac{u^{n}_{i} - u^{n}_{i-1}}{\Delta x}, $$

and see what effect it has on the solution error.

```python
# New sub-routine for evaluating the RHS function

def funcO1(u,epsilon):
    # Sub-routine to evaluate the RHS function that uses a 2nd-order centered differences to discretize the diffusion term
    # and 1st-order backwards difference for the convection term
    
    f = 0*u.copy()
    f[0] = epsilon*(u[1]-2*u[0]+u[-2])/(dt**2) - (u[0]-u[-2])/dx
    f[1:-1] = epsilon*(u[2:]-2*u[1:-1]+u[:-2])/(dt**2) - (u[1:-1]-u[:-2])/dx
    f[-1] = epsilon*(u[1]-2*u[-1]+u[-2])/(dt**2) - (u[-1]-u[-2])/dx
    return f 
```

```python
u = np.exp(-x**2)
plt.plot(x,u,label='$t=0$')

tmax = 20
epsilon = 0
t = 0
n = 0
while (t < tmax):
    un = u.copy()
    u1 = un + 0.5*dt*funcO1(un,epsilon) # Step 1
    u2 = un + 0.5*dt*funcO1(u1,epsilon) # Step 2
    u3 = un + dt*funcO1(u2,epsilon)     # Step 3
    u = un + (1.0/6.0)*dt*(funcO1(un,epsilon) + 2.0*funcO1(u1,epsilon) + 2.0*funcO1(u2,epsilon) + funcO1(u3,epsilon))  # Step 4
    t+=dt

plt.plot(x,u,label='$t = 20$')
plt.title('Runge-Kutta, $\epsilon = 0$')
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$u$')
plt.legend()
print('Undamped L2norm(error) =',L2norm(u))
```

We see that the phase error has been eliminated, but the first-order backwards difference suffers from large dissipation error. After only 20 s (2 cycles), the wave amplitude has diminished to below 50% of the original height. Note that this is despite have a dampening coefficient of $\epsilon = 0$--the dissipation is entirely due to numerical error.
